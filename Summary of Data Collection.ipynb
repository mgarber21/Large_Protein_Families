```python
import pandas as pd
import os
```

```python
dict_ranks = dict()
dict_num_sp = dict()

for file in os.listdir(os.path.join(os.getcwd(), 'SampledGenomes')):
    if ".csv" in file:
        Dir = os.path.join(os.getcwd(), 'SampledGenomes', file) 
        # print(Dir)
        df = pd.read_csv(Dir)
        ranks = [ 'superkingdom', 'phylum',
           'class', 'order', 'family', 'genus', 'species']


        for i, rank in enumerate(ranks):
            if len(df[rank].unique()) > 2:
                rank_below = rank
                rank = ranks[i-1]
                break
            else:
                continue
                
        number_of_sp = []
        for i in df[rank_below]:
            number_of_sp.append(len(df[df[rank_below] == i]["species"]))
        dict_ranks[file.replace("Sampled", "").replace(".csv", "")] = [len(df[rank_below].unique()),max(number_of_sp)]
        # dict_num_sp[file.replace("Sampled", "").replace(".csv", "")] = 
# 
        
```

```python
pd.DataFrame.from_dict(dict_ranks).T.rename(columns = dict(zip([0, 1], ["unique_taxa_in_rank_below", "max_sp_per_rank_below"]))).sort_index()
```
